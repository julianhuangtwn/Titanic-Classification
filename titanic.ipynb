```python
# This 
import numpy as np 
import pandas as pd 
```

```python
# read_csv reads data, and return a 2D data frame object
train_data = pd.read_csv("train.csv")
# print(train_data)
# .head by default displays first 5 records
train_data.head()
```

```python
#Loads test data
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
#Creates a panda series that looks for all females, and survival status
women = train_data.loc[train_data.Sex == 'female',"Survived"]
women_alive_rate = sum(women)/len(women)

print("% of women who survived: ", women_alive_rate)
```

```python
men = train_data.loc[train_data.Sex == 'male', "Survived"]
men_alive_rate = sum(men)/len(men)

print("% of men who survived: ", men_alive_rate)
# Women survival rate 0.74, while men is only 0.19, shows that gender was an important factor of prediction
```

```python
from sklearn.ensemble import RandomForestClassifier

# y is the prediction. We want to know who survived or not
y = train_data["Survived"]

# features are the variables we use to train the model, which we think affect the prediction
features = ["Pclass", "Sex", "SibSp", "Parch"]

# One Hot Encoding, however, Pclass is ordinal, Sex is nominal, but SibSp (Sibling and Spouse) and Parch (Parent and Child) are discrete
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('prediction.csv', index=False)

```
